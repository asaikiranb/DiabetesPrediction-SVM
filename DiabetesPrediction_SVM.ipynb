Importing the dependencies


```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Dataset loading - The dataset used in this project primarily consists of health-related data from female patients, focusing on attributes such as the number of pregnancies, glucose levels, blood pressure, skin thickness, insulin levels, BMI, diabetes pedigree function, and age.

```python
#loading the dataset
```

```python
diabetes_dataset= pd.read_csv('/dev/diabetes.csv')
diabetes_dataset.head()
```

```python
#number of rows and column in the dataset
diabetes_dataset.shape
```

Statistical mesasures related to this dataset


```python
diabetes_dataset.describe()
```

```python
# number of people with diabetes and without diabetes in this dataset
diabetes_dataset['Outcome'].value_counts()
```

The target variable, 'Outcome', is labeled as 0 for non-diabetic and 1 for diabetic patients.


```python
# mean value for diabetic and non diabetic case
diabetes_dataset.groupby('Outcome').mean()
```

```python
# saving outcome and data in seperate variables, since we are droping a column the axis is 1
X=diabetes_dataset.drop(columns='Outcome', axis =1)
Y=diabetes_dataset['Outcome']
print(X)
print(Y)
```

Standardizing the data values to a uniform range so that it increases model accuracy and learning efficiency.

```python
scaler= StandardScaler()
scaler.fit(X)
standardized_data= scaler.transform(X)
print(standardized_data)
# so now all values will be in similar range
```

```python
X=standardized_data
y= diabetes_dataset['Outcome']
print(X)
print(Y)
```

Splitting the data into train and test

```python
X_train,X_test , Y_train, Y_test = train_test_split(X,Y, train_size=0.2, stratify=Y, random_state=2)
# We are stratifying using y, our output variable, to ensure we have equal diabetic and non-diabetic cases in each split of train and test.
# The random_state is used to replicate the splitting process across different systems.
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Training SVM model


```python
## Initialize the Support Vector Machine classifier with a linear kernel
classifier = svm.SVC(kernel='linear')
classifier.fit(X_train,Y_train)
```

Model evaluation

```python
# accuracy score from the training data
X_train_prediction= classifier.predict(X_train)
training_data_accuracy= accuracy_score(X_train_prediction,Y_train)
print(training_data_accuracy)
```

```python
# accuracy score from the testing data
X_test_prediction= classifier.predict(X_test)
test_data_accuracy= accuracy_score(X_test_prediction,Y_test)
print(test_data_accuracy)
```

As expected accuracy score from the testing data < accuracy score from the training data
as testing data is never seen by our model


Making a Predictive System


```python
input_data = (5,166,72,19,175,25.8,0.587,51)

# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardize the input data
std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
